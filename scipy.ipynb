```python
import numpy as np
import scipy as sp
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
%matplotlib inline
```

### Otimização

```python
def rosen(x):
    """The Rosenbrock function"""
    return sum(100.0*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0)
```

```python
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
res = minimize(rosen, x0, method='nelder-mead',
               options={'xtol': 1e-8, 'disp': True})
```

#### Distribuições em estatística

```python
from scipy.stats import norm

fig, ax = plt.subplots(1, 2, figsize=(14,6))

xnor = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
ax[0].plot(xnor, norm.pdf(xnor),
       'r-', lw=5, alpha=0.6, label='norm pdf')

from scipy.stats import chi
df = 2
xchi = np.linspace(chi.ppf(0.01, df),
                chi.ppf(0.99, df), 100)
ax[0].plot(xchi, chi.pdf(xchi, df),
       'g--', lw=5, alpha=0.6, label='Chi square pdf')


# =============
# cdf
# =============

ax[1].plot(xnor, norm.cdf(xnor),
       'r-', lw=5, alpha=0.6, label='norm pdf')

ax[1].plot(xchi, chi.cdf(xchi,df),
       'g--', lw=5, alpha=0.6, label='norm pdf')
fig.legend(bbox_to_anchor=(1.5,1))
```

```python
import statsmodels.api as sm

ax = sm.qqplot(xchi, line='q')
```
